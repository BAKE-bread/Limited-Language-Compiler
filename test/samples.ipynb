```cpp
program Fibonacci {
   func fib(n) {
       let a = 0;
       let b = 1;
       let temp;
       let count = 0;
       while (count < n) {
           output(b);
           temp = a + b;
           a = b;
           b = temp;
           count = count + 1;
       };
       return a;
   }
   main {
       let num;
       input(num);
       let x = fib(num);
       output(x); // Output the final result of fib(num)
   }
}
```

```cpp
program GCD {
    func gcd(a, b) {
        while (b != 0) {
            let quotient = a / b;
            let temp = b;
            b = a - quotient * b;
            a = temp;
        };
        return a;
    }
    main {
        let x;
        let y;
        input(x, y);
        output(gcd(x, y));
    }
}
```

```cpp
program Factorial {
    func factorial(n) {
        let result;
        if (n == 0) {
            result = 1;
        } else {
            let temp = factorial(n-1);
            result = n * temp;
        };
        return result;
    }
    main {
        let num;
        input(num);
        let x = factorial(num);
        output(x);
    }
}
```

```cpp
program ChainedCallTest {
    // 函数1: 一个简单的工具函数，给输入值增加一个固定的偏移
    func add_offset(val: int) {
        let result;
        result = val + 100;
        return result; 
    }

    // 函数2
    func calculate_series(base: int, multiplier: int) {
        let temp_val;
        let final_val;
        // 调用函数1
        temp_val = add_offset(base); 
        // 在函数1的结果上进行进一步计算
        final_val = temp_val * multiplier;
        return final_val;
    }

    // 函数3
    func high_level_process(start_val: int) {
        let series_result;
        // 调用函数2
        series_result = calculate_series(start_val, 2); 
        return series_result;
    }

    main {
        let initial_val = 5;
        let result_from_1;
        let result_from_3;

        result_from_1 = add_offset(initial_val); 
        // 预期输出: 105 (5 + 100)
        output(result_from_1);

        result_from_3 = calculate_series(initial_val, 3);
        // Main 直接调用函数3, 内部会调用到函数1
        // 预期输出: 315
        output(result_from_3);
    }
}
```

```cpp
program Shadowing {
    func f(a, b) {
        let result = 0;
        if (a > b) {
            let result = 10;    // 遮蔽外部result
            let c = a * b;      // 内部作用域变量
            output(result, c);  // 10, 15
        } else {
            let result = 20;    // 遮蔽外部result
            let c = a + b;      // 内部作用域变量
            output(result, c);  // 20, 8
        };
        result = a + b;         // 8
        output(result);

        while (b > 0) {         // 嵌套作用域示例
            b = b - 1;  
            let a = b * 2;      // 遮蔽参数a
            output(a);          // 4 2 0
            output(b);          // 2 1 0 
        };
        
        a = 7;
        let x = a;
        return x;               // 最后一句return 7
    }
    main {
        let num = 6; 
        let i = 0;
        output(num);            // 6
        if (i < 1)
        {
            let num = -8 * ((4 + 3) / 5 - 10);   // 遮蔽外部num
            output(num);        // 72 =  8 * (10-1)
        };
        output(num);            // 外部num 6
        let x = 5;
        let y = 3;
        output(x, y);           // 5, 3
        let result = f(x, y);
        output(x, y);           // 5, 3
        output(result);         // 7
    }
}
```

```cpp
program StudentGrades {
    struct Student {
        int id;
        int scores[3];
    };

    func calculateAverage(s: struct Student) {
        let sum = 0;
        let count = 3; // Assuming always 3 scores for this example
        let avg = 0;
        let i = 0;

        while (i < 3) {
            sum = sum + s.scores[i];
            i = i + 1;
        };

        if (count > 0) {
            avg = sum / count;
        };
        // output(avg); // For debugging inside function
        return avg;
    }

    main {
        let students: struct Student[2];
        let avg_score = 0;

        students[0].id = 101;
        students[0].scores[0] = 85;
        students[0].scores[1] = 90;
        students[0].scores[2] = 78; // avg = (85+90+78)/3 = 253/3 = 84

        students[1].id = 102;
        students[1].scores[0] = 92;
        students[1].scores[1] = 88;
        students[1].scores[2] = 95; // avg = (92+88+95)/3 = 275/3 = 91

        avg_score = calculateAverage(students[0]);
        output(students[0].id);
        output(avg_score); // Should be 84

        avg_score = calculateAverage(students[1]);
        output(students[1].id);
        output(avg_score); // Should be 91
    }
}
```

```cpp
program BubbleSortTest {
    // 由于不支持直接传递数组，我们将其封装在结构体中
    // 注意：数组大小是固定的
    struct IntArrayContainer {
        int data[8];
    };

    // 函数1: 打印数组内容
    // 接收一个包含数组的结构体作为参数
    func print_array(arr_container: struct IntArrayContainer) {
        let i = 0;
        output(99999); // 输出一个分隔符，便于观察
        while (i < 8) {
            output(arr_container.data[i]);
            i = i + 1;
        };
        return 0;
    }

    // 函数2: 对数组进行冒泡排序
    // 同样接收一个包含数组的结构体作为参数
    func bubble_sort(arr_container: struct IntArrayContainer) {
        let i = 0;
        let j;
        let temp;
        let n = 8;

        while (i < n - 1) {
            j = 0;
            while (j < n - i - 1) {
                if (arr_container.data[j] > arr_container.data[j+1]) {
                    // 交换元素
                    temp = arr_container.data[j];
                    arr_container.data[j] = arr_container.data[j+1];
                    arr_container.data[j+1] = temp;
                };
                j = j + 1;
            };
            i = i + 1;
        };

        // 在函数内部打印排序后的数组，以证明排序已在副本上完成
        output(88888); // 输出一个分隔符
        let k = 0;
        while (k < n) {
            output(arr_container.data[k]);
            k = k + 1;
        };

        return 0; // 函数必须有返回值
    }


    main {
        let my_array: struct IntArrayContainer;

        // 初始化数组数据
        my_array.data[0] = 64;
        my_array.data[1] = 34;
        my_array.data[2] = 25;
        my_array.data[3] = 12;
        my_array.data[4] = 22;
        my_array.data[5] = 11;
        my_array.data[6] = 90;
        my_array.data[7] = 5;

        // 1. 打印原始数组
        // 预期输出: 99999, 64, 34, 25, 12, 22, 11, 90, 5
        print_array(my_array);

        // 2. 调用冒泡排序函数
        // 函数内部会打印排序后的结果
        // 预期输出: 88888, 5, 11, 12, 22, 25, 34, 64, 90
        bubble_sort(my_array);

        // 3. 再次打印原始数组
        // 由于结构体是值传递，main中的 my_array 应该保持不变
        // 预期输出: 99999, 64, 34, 25, 12, 22, 11, 90, 5
        print_array(my_array);
    }
}
```
